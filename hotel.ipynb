```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_combined['hotel_info'])
```

```python
def recommend_hotels_based_on_query(query, df, top_n=3):
    # Add the user query to the dataset for vectorization
    data_with_query = df_combined['hotel_info'].tolist() + [query]

    # Vectorize the combined text fields and the query
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(data_with_query)

    # Compute cosine similarity between the query and all hotels
    cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

    # Add similarity scores to the DataFrame
    df['similarity_score'] = cosine_similarities

    # Sort by similarity score and get top N recommendations
    recommendations = df.sort_values(by='similarity_score', ascending=False).head(top_n)

    return recommendations[['property_name', 'similarity_score', 'guest_recommendation', 'address']]

# Example usage: Recommend hotels based on a user query
query = "I need swimming pool"
recommendations = recommend_hotels_based_on_query(query, df_combined)
print(recommendations)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

def find_cluster(latitude, longitude):
    # Fit NearestNeighbors to find the closest cluster to user's location
    cluster_centroids = hotels_df.groupby('cluster')[['latitude', 'longitude']].mean()
    neighbors = NearestNeighbors(n_neighbors=1)
    neighbors.fit(cluster_centroids)
    distance, nearest_cluster = neighbors.kneighbors([[latitude, longitude]])

    # Return the cluster ID of the closest cluster
    return cluster_centroids.index[nearest_cluster[0][0]]

# User location
user_latitude, user_longitude = 32.225, 76.327
target_cluster = find_cluster(user_latitude, user_longitude)

# Filter hotels by target cluster
cluster_df = df_combined[df_combined['cluster'] == target_cluster]

cluster_df
```

```python
def recommend_hotels_based_on_query(query, cluster_df, top_n=3):
    # Append the user query to the DataFrame for TF-IDF processing
    data_with_query = cluster_df['hotel_info'].tolist() + [query]

    # Vectorize the text fields
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(data_with_query)

    # Compute cosine similarity between query and hotels in the cluster
    cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

    # Add similarity scores to the DataFrame
    cluster_df = cluster_df.copy()  # Avoid modifying original
    cluster_df['similarity_score'] = cosine_similarities

    # Sort by similarity score and return top N results
    recommendations = cluster_df.sort_values(by='similarity_score', ascending=False).head(top_n)

    return recommendations[['property_name', 'similarity_score', 'guest_recommendation', 'address']]

# User's query for hotel recommendation
query = "swimming pool Internet"
recommendations = recommend_hotels_based_on_query(query, cluster_df)
print(recommendations)

```
