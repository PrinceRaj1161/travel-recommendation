```python
# Function to filter hotels based on user preferences
def filter_hotels(cluster, cuisine, max_cost):
    filtered_hotels = df[
        (df['cluster'] == cluster) &
        (df['Cuisines'].str.contains(cuisine, case=False)) &
        (df['Average Cost for two'] <= max_cost)
    ]
    return filtered_hotels

# Function to sort hotels based on ratings and cost
def sort_hotels(hotels_df):
    return df.sort_values(by=['rating_integer', 'Average Cost for two'], ascending=[False, True])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors



def find_cluster(latitude, longitude):
    cluster_centroids = df.groupby('cluster')[['Latitude', 'Longitude']].mean()
    neighbors = NearestNeighbors(n_neighbors=1)
    neighbors.fit(cluster_centroids)
    _, nearest_cluster = neighbors.kneighbors([[latitude, longitude]])
    return cluster_centroids.index[nearest_cluster[0][0]]
```

```python
# Function to recommend based on cuisine query within a specific cluster
def recommend_restaurants(cuisine_query, cluster_df, top_n=3):
    cluster_df = cluster_df.copy()
    vectorizer = TfidfVectorizer()

    # Create TF-IDF matrix with cuisines and user query
    cuisine_data = cluster_df['Cuisines'].tolist() + [cuisine_query]
    tfidf_matrix = vectorizer.fit_transform(cuisine_data)
    cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

    cluster_df['similarity_score'] = cosine_similarities

    # Calculate a weighted score using rating, cost, and similarity
    cluster_df['weighted_score'] = (
        cluster_df['similarity_score'] * 0.5 +
        (cluster_df['rating_integer'] / 5) * 0.3 -
        (cluster_df['Average Cost for two'] / cluster_df['Average Cost for two'].max()) * 0.2
    )

    recommendations = cluster_df.sort_values(by='weighted_score', ascending=False).head(top_n)
    return recommendations[['Restaurant Name', 'similarity_score', 'rating_integer', 'Average Cost for two']].to_dict(orient='records')
```

```python
# Main function to take inputs and provide recommendations
def get_recommendations(latitude, longitude, cuisine_query):
    # Step 1: Predict the cluster based on latitude and longitude
    target_cluster = find_cluster(latitude, longitude)

    # Step 2: Filter restaurants in the target cluster
    cluster_df = df[df['cluster'] == target_cluster]

    # Step 3: Get recommendations
    recommendations = recommend_restaurants(cuisine_query, cluster_df)
    return recommendations
```

```python
# Example Usage
latitude_input = 32.220
longitude_input = 76.322
cuisine_query_input = "Italian"

recommendations = get_recommendations(latitude_input, longitude_input, cuisine_query_input)
print("Recommended Restaurants:")
recommendations
```
